```python
print("\n'Welcome to the PyLogic Suite: Integrated Pyramid Generator & Number Analyzer'".upper())
while True :
    
    print("\nChoose Your Number: ")
    print("\n1. Pattern Generator")
    print("2. Number Analyzer")
    print("3. Know You Birth Year")
    print("4. Exit")
    
    user = int(input("\nEnter Your Number"))
    
    if user == 1:
        print("\nWelcome to Pattern Generator:".upper())

        n = 1
        Number = int(input("Enter Row For Pyramid: "))

        for i in range (1 , Number+1):
            print(" " * (Number - i) + "*" * (2 * i - 1))
              
    elif user == 2:
        print("\nWelcome to the Number Analyzer".upper())

        Number_1 = int(input("Enter Your Number 1: "))
        Number_2 = int(input("Enter Your Number 2: "))
        print(f"\nODD and EVEN Numbers between {Number_1} and {Number_2}".upper())
        print("\n")
        for i in range (Number_1 , Number_2+1):
            if i % 2 == 0:
                print(f"{i}: EVEN Number")
            else:
                print(f"{i}: ODD Number")
    
    elif user == 3:
        print("\nKnow Your Birth Year By Your Age".upper())

        age = int(input("Enter Your Age"))

        Birth_Year = 2026 - age

        print(f"\n'Your Birth Year Is {Birth_Year}, 32 Based On Your Age!'")

    elif user == 4:
        print("\nThank You! For Using This Program, Goodbye!")
        break

    elif user >= 4:
        print(f"Sorry, Your Number {user} Is Not In Option.")
```
